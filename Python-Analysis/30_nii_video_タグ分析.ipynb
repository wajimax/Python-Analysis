```python
#交差検証
%matplotlib inline
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import defaultdict
import codecs
import seaborn as sns
```

```python
video_tsv_merge_data = pd.read_table('./Output/video_tsv_merge_data.tsv',sep="\t", index_col=None)
```

```python
# UNIX時間を変換
video_tsv_merge_data['upload_time'] =  pd.to_datetime(video_tsv_merge_data['upload_time'], unit='s')
video_tsv_merge_data['year'] = video_tsv_merge_data['upload_time'].dt.year
video_tsv_merge_data['month'] = video_tsv_merge_data['upload_time'].dt.month
video_tsv_merge_data['day'] = video_tsv_merge_data['upload_time'].dt.day
video_tsv_merge_data['hour'] = video_tsv_merge_data['upload_time'].dt.hour
video_tsv_merge_data['weekday_name'] = video_tsv_merge_data['upload_time'].dt.weekday_name
```

```python
video_tsv_merge_data = video_tsv_merge_data.fillna({'category':'category_NaN'})
video_tsv_merge_data
```

```python
#year
video_tsv_merge_pivot = pd.pivot_table(video_tsv_merge_data, index=['category'],columns=['year'], values=['watch_num'],aggfunc=np.sum, dropna=True)
video_tsv_merge_pivot = video_tsv_merge_pivot.fillna(0)
video_tsv_merge_pivot.to_csv('./OutPut/pivot_data/year/video_pivot_category_watch_num.tsv', sep='\t')

video_tsv_merge_pivot = pd.pivot_table(video_tsv_merge_data, index=['category'],columns=['year'], values=['comment_num'],aggfunc=np.sum, dropna=True)
video_tsv_merge_pivot = video_tsv_merge_pivot.fillna(0)
video_tsv_merge_pivot.to_csv('./OutPut/pivot_data/year/video_pivot_category_comment_num.tsv', sep='\t')

video_tsv_merge_pivot = pd.pivot_table(video_tsv_merge_data, index=['category'],columns=['year'], values=['mylist_num'],aggfunc=np.sum, dropna=True)
video_tsv_merge_pivot = video_tsv_merge_pivot.fillna(0)
video_tsv_merge_pivot.to_csv('./OutPut/pivot_data/year/video_pivot_category_mylist_num.tsv', sep='\t')

video_tsv_merge_pivot = pd.pivot_table(video_tsv_merge_data, index=['category'],columns=['year'], values=['comment_num'],aggfunc=np.sum, dropna=True)
video_tsv_merge_pivot = video_tsv_merge_pivot.fillna(0)
video_tsv_merge_pivot.to_csv('./OutPut/pivot_data/year/video_pivot_category_comment_num.tsv', sep='\t')
```
